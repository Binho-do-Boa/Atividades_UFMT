### Pacotes necessários

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from tensorflow import keras
```

### Dados

```
notas = pd.read_csv("notas.csv")
notas.head()
```

```
notas.describe()
```

```
x = notas.iloc[:, 0:3].values
y = notas.iloc[:, 3].values
```

### Definição do modelo

Adição de camadas, atribuindo neurônios e definindo como será o formato da entrada

```
model = keras.Sequential()
model.add(keras.layers.Dense(7, input_dim=3, activation='tanh'))
model.add(keras.layers.Dense(3, input_dim=3, activation='tanh'))
model.add(keras.layers.Dense(1, activation='linear'))
```

loss: avaliação de como estão as estimativas

optimizer: forma de corrigir os pesos baseado no loss

```
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['mean_absolute_error']) #mean_absolute_error
```

### Treinamento

```
history = model.fit(x, y, epochs=100)
```

```
plt.plot(history.history['mean_absolute_error'])
plt.show()
```

### Estimando novos valores

```
result = model.predict(np.array([
                        [0,0,0],
                        [0.75,0,0],
                        [0.5,0,0.5],
                        [1, 1, 1]
                       ]))
print(result)
```
